```python
from typing_extensions import TypedDict
from langgraph.graph import StateGraph
```

```python
class MyState(TypedDict): # state schema
  name: str

def appreciate_user(state: MyState) -> MyState:
  """Appreciate the user for learning LangGraph."""
  
  state['name'] = state['name'] + ", you are doing a amazing job learning LangGraph!"

  return state
```

```python
graph = StateGraph(MyState)

graph.add_node("appreciater", appreciate_user)

graph.set_entry_point("appreciater")
graph.set_finish_point("appreciater")

app = graph.compile()
app
```

```python
result = app.invoke({"name": "Bob"})
result['name']
```
